```python
# Import the dependencies
import pandas as pd
import numpy as np
import gmaps
import requests
# Import the API key
from config import g_key
```

```python
# Configure gmaps to use your Google API key
gmaps.configure(api_key=g_key)
```

```python
# Store the WeatherPy_database.csv into a DataFrame
vacay_data_df = pd.read_csv("weather_data/WeatherPy_vacation.csv")
# Drop null values
vacay_df = vacay_data_df.dropna()
# Display DataFrame
vacay_df.head()
```

```python
# Verify NO null values
vacay_df.isnull().sum()
```

```python
# List of countries to choose from:
countries_list = vacay_df["Country"].unique()
# Display list of countries to choose from:
countries_list
```

```python
# Using the countries_list, input which country you would like to visit?
country = str(input("Based on countries list, which country would you like to visit? "))
```

```python
# Filter the dataset based on the country you would like to visit
vacay_cities_df = vacay_df.loc[(vacay_df["Country"] == country)]
vacay_cities_df
```

```python
# Convert the DataFrame to a NumPy array
cities = vacay_cities_df.to_numpy()
```

```python
# Choose 4 cities you wish to visit
# Note: city names are at index[1]
city1 = cities[4][1]
city2 = cities[17][1]
city3 = cities[14][1]
city4 = cities[28][1]
print(city1, city2, city3, city4)
```

```python
# Assign a variable to the country of each city
# Note: Country is at index[2]
city1_country = cities[4][2]
city2_country = cities[17][2]
city3_country = cities[14][2]
city4_country = cities[28][2]
print(city1_country, city2_country, city3_country, city4_country)
```

```python
# Assign a variable to the current weather description of each city
# Note: Current Description is at index[4]
city1_weather = cities[4][4]
city2_weather = cities[17][4]
city3_weather = cities[14][4]
city4_weather = cities[28][4]
print(city1_weather, city2_weather, city3_weather, city4_weather)
```

```python
# Assign a variable to the latitude of each city
# Note: Latitude is at index[5]
city1_lat = cities[4][5]
city2_lat = cities[17][5]
city3_lat = cities[14][5]
city4_lat = cities[28][5]
print(city1_lat, city2_lat, city3_lat, city4_lat)
```

```python
# Assign a variable to the longitude of each city
# Note: Longitude is at index[6]
city1_lng = cities[4][6]
city2_lng = cities[17][6]
city3_lng = cities[14][6]
city4_lng = cities[28][6]
print(city1_lng, city2_lng, city3_lng, city4_lng)
```

```python
# Assign a variable to the hotel name being visited in each city
# Note: Hotel Name is at index[7]
city1_hotel = cities[4][7]
city2_hotel = cities[17][7]
city3_hotel = cities[14][7]
city4_hotel = cities[28][7]
print(city1_hotel, city2_hotel, city3_hotel, city4_hotel)
```

```python
# Pair each latitude and longitude of each city
city1_coord = (city1_lat, city1_lng)
city2_coord = (city2_lat, city2_lng)
city3_coord = (city3_lat, city3_lng)
city4_coord = (city4_lat, city4_lng)
```

```python
# Latitude-Longitude Pairs
lat_lng_1 = (city1_coord)
lat_lng_2 = (city2_coord)
lat_lng_3 = (city3_coord)
lat_lng_4 = (city4_coord)

# Create a gmaps figure
fig = gmaps.figure()

# Add directions from city1 to city4 and pass through city2 and city3
directions_layer = gmaps.directions_layer(lat_lng_1, lat_lng_4,
                                          waypoints=[lat_lng_2, lat_lng_3], 
                                          show_markers=True,
                                          travel_mode='DRIVING')

# Add the directions_layer to the figure
fig.add_layer(directions_layer)

# Call the figure
fig
```

```python
# Save each city data into a DataFrame
city1_df = pd.DataFrame([{"City" : city1,
            "Country" : city1_country,
            "Current Description" : city1_weather,
            "Latitude" : city1_lat,
            "Longitude" : city1_lng, 
            "Hotel Name" : city1_hotel}])
city2_df = pd.DataFrame([{"City" : city2,
            "Country" : city2_country,
            "Current Description" : city2_weather,
            "Latitude" : city2_lat,
            "Longitude" : city2_lng, 
            "Hotel Name" : city2_hotel}])
city3_df = pd.DataFrame([{"City" : city3,
            "Country" : city3_country,
            "Current Description" : city3_weather,
            "Latitude" : city3_lat,
            "Longitude" : city3_lng, 
            "Hotel Name" : city3_hotel}])
city4_df = pd.DataFrame([{"City" : city4,
            "Country" : city4_country,
            "Current Description" : city4_weather,
            "Latitude" : city4_lat,
            "Longitude" : city4_lng, 
            "Hotel Name" : city4_hotel}])
```

```python
# Combine the DataFrames
city_choices_df = pd.concat([city1_df, city2_df, city3_df, city4_df])
city_choices_df
```

```python
# info_box_template variable
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Description}</dd>
</dl>
"""
```

```python
# Store the DataFrame Row
city_info = [info_box_template.format(**row) for index, row in city_choices_df.iterrows()]
```

```python
# Add a marker for each location and info box
locations = city_choices_df[["Latitude", "Longitude"]]

# Latitude-Longitude Pairs
lat_lng_1 = (city1_coord)
lat_lng_2 = (city2_coord)
lat_lng_3 = (city3_coord)
lat_lng_4 = (city4_coord)

# Create a gmaps figure
fig = gmaps.figure()

# Add directions from city1 to city4 and pass through city2 and city3
directions_layer = gmaps.directions_layer(lat_lng_1, lat_lng_4,
                                          waypoints=[lat_lng_2, lat_lng_3], 
                                          show_markers=False,
                                          travel_mode='DRIVING')
# Add a marker for each location and info box
marker_layer = gmaps.marker_layer(locations, info_box_content=city_info)

# Add the directions_layer to the figure
fig.add_layer(directions_layer)
# Add the marker_layer to the figure
fig.add_layer(marker_layer)

# Call the figure to plot the data
fig
```
